```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris;
```

```python
iris = load_iris();
features = list(map(lambda s : s[:7],iris["feature_names"]));
irisDF = pd.DataFrame(iris.data,columns=features);
irisDF['type'] = iris["target_names"][iris["target"]];
irisDF = irisDF.sample(frac=1);
irisDF.head()
```

# Bar chart

```python
barData = irisDF.iloc[35:55];
plt.figure(figsize=(20,10))
plt.subplot(1,2,1);
sns.barplot(x=barData["sepal l"], y = barData["petal l"],hue=barData["type"]);

plt.subplot(1,2,2);
sns.barplot(x=barData["sepal w"], y = barData["petal w"],hue=barData["type"]);
```

## Count Bar 

```python
countData = irisDF.iloc[:50];
plt.figure(figsize=(20,10))
plt.subplot(1,2,1);
sns.countplot(x=countData["sepal l"],hue=countData["type"]);

plt.subplot(1,2,2);
sns.countplot(x=countData["sepal w"],hue=countData["type"]);
```

## FacetGrid

```python
g = sns.FacetGrid(irisDF,col='type');
g.map(plt.scatter, "sepal l", "petal l");
g.add_legend()
```

## swarmplot

```python
plt.figure(figsize=(15,10))
plt.subplot(221);
sns.swarmplot(x='type',y='sepal l',data=irisDF)

plt.subplot(222);
sns.swarmplot(x='type',y='sepal w',data=irisDF)

plt.subplot(223);
sns.swarmplot(x='type',y='petal l',data=irisDF)

plt.subplot(224);
sns.swarmplot(x='type',y='petal w',data=irisDF)
```

## BoxPlot

```python
sns.boxplot(x='type', y='petal l', data=irisDF)
```
