```python
import numpy as np
import pandas as pd
import seaborn
```

```python
from sklearn import preprocessing
preprocessing.
X = np.float64([[2,10],[4,2],[6,17],[18,9],[10,11]])
print("Original X : \n",X)
# Generalize the data by making mean to 0 and variance to 1
sscaler = preprocessing.StandardScaler().fit_transform(X)
print("StandardScaler X : \n",sscaler)
print("Mean for first {} second {} column ".format(np.mean(sscaler[:,0]),np.mean(sscaler[:,1])))
print("Variance for first {} second {} column ".format(np.var(sscaler[:,0]),np.var(sscaler[:,1])))
```

```python
# Generalize the data by keeping all values b/w 0 and 1. (x - xmin)/(xmax - xmin)
minMaxScaler = preprocessing.MinMaxScaler().fit_transform(X)
print("MinMaxScaler X : \n",minMaxScaler)
```

```python
robustScaler = preprocessing.RobustScaler().fit_transform(X)
print("RobustScaler X : \n",robustScaler)
```

```python
quantileScaler = preprocessing.QuantileTransformer().fit_transform(X)
print("RobustScaler X : \n",quantileScaler)
```

```python
seaborn.lineplot(x="x",y="y",data=pd.DataFrame({"x":X[:,0],"y":X[:,1]}),label="X")
seaborn.lineplot(x="x",y="y",data=pd.DataFrame({"x":sscaler[:,0],"y":sscaler[:,1]}),label="StandardScaler")
seaborn.lineplot(x="x",y="y",data=pd.DataFrame({"x":minMaxScaler[:,0],"y":minMaxScaler[:,1]}),label="MinMaxScaler")
seaborn.lineplot(x="x",y="y",data=pd.DataFrame({"x":robustScaler[:,0],"y":robustScaler[:,1]}),label="RobustScaler",color='black')
seaborn.lineplot(x="x",y="y",data=pd.DataFrame({"x":quantileScaler[:,0],"y":quantileScaler[:,1]}),label="QuantileTransformer")
```
