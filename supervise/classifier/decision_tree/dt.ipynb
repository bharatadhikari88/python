```python
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris;
import graphviz
```

```python
iris = load_iris();
iris.data[:5]
```

### Gini index and entropy(information gain) are two feature prediction technique for decision tree classification After feature selection, tree is sub-divided into discrete values if feature has discrete values else mean of concrete value is used for split

```python
trainX,testX,trainy,testy = train_test_split(iris.data,iris.target,random_state=5);
clf = DecisionTreeClassifier(criterion='gini',max_depth=3).fit(trainX,trainy);
clf.score(testX,testy)
```

### Write decision tree into file

```python
export_graphviz(clf,out_file='iris_decision_tree.dot',class_names=iris.target_names,feature_names=iris.feature_names);
```

### read dot file

```python
with open('iris_decision_tree.dot','r') as f:
    display(graphviz.Source(f.read()));
```
