## SVC :- Support vector classifer
svc separate the data either by drawing line, oval, polynomial depends on type of kernal
provided to it.
### It use ovo(one vs one).
If target has 4 possible values. kernal=linear will create (n-1)! lines (1 vs 2, 1 vs 3, 1 vs 4, 2 vs 3, 2 vs 4, 3 vs 4)  

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import mglearn
```

```python
data = load_iris();
irisDF = pd.DataFrame(data.data,columns=data.feature_names);
irisDF["type"] = data.target;
irisDF.head()
```

```python
trainX,testX,trainy,testy = train_test_split(data.data[:,0:2],data.target,random_state=6);
def plotClassifier(classifier):
    plt.figure(figsize=(15,5));
    plt.xlabel('sepal length (cm)');
    plt.ylabel('sepal width (cm)');
    markers = {'0': "o", '1': "s",2:"X"}
    sns.scatterplot(data=irisDF,x='sepal length (cm)',y='sepal width (cm)',hue='type');
    mglearn.plots.plot_2d_classification(classifier,data.data[:,0:2],fill=True,alpha=0.7);
```

```python
## liner SVC classifier
plotClassifier(LinearSVC(max_iter=10000).fit(trainX,trainy));
```

```python
## linear classifier
plotClassifier(SVC(kernel='linear',gamma=1).fit(trainX,trainy));
```

```python
## rbf classifier
plotClassifier(SVC(kernel='rbf',gamma=1).fit(trainX,trainy));
```

```python
## poly classifier
plotClassifier(SVC(kernel='poly',gamma=1).fit(trainX,trainy));
```
